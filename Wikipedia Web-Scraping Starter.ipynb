### Wiki - Web Scrapping PoC
This is a sample proof of concept which demonstrates usage of Wiki API to fetch categories and respective sub-categories upto n levels. This example could be extended as per the requirements.

```python
import requests
import json

def get_category(starting_cat, params = None):
    """
    Sample Code!
    Gets subcategories for a given starting category. 
    Use configurable params option in PROD.
    """
    session = requests.Session()

    url = "https://en.wikipedia.org/w/api.php"
    
    if params is None:
        params = {
            "action": "query",
            "cmtitle": starting_cat,
            "cmlimit": "5",
            "list": "categorymembers",
            "format": "json"
        }

    req = session.get(url=URL, params=params)
    d = req.json()

    return d['query']['categorymembers']
    

def get_categories(starting_cat, data, level=2):
    """
    Sample Code!
    Recursively constructs dict of categories and their respective sub categories upto specified level.
    """
    if level == 0:
        return data
    else:
        sub_categories = get_category(starting_cat)
        for sub_cat in sub_categories:
            if starting_cat in data:
                data[starting_cat][sub_cat['title']] = {'pageid': sub_cat['pageid']}
            else:
                data[starting_cat] = {}
                data[starting_cat][sub_cat['title']] = {'pageid': sub_cat['pageid']}
            get_categories(sub_cat['title'], data[starting_cat], level-1)
    
```

```python
cat = 'Category:Companies_of_the_United_States_by_state_and_city'
r = {}
get_categories(cat, r, 3)
```
